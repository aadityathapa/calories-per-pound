```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
# Total calories & protein
df['Total calories'] = ((df['Calories per 100g'] / 100) * df['Weight (g)']).round().astype(int)
df['Total protien'] = ((df['Protein per 100g'] / 100) * df['Weight (g)']).round().astype(int)

# Value metrics
df['Calories per pound'] = (df['Total calories'] / df['Price (£)']).round().astype(int)
df['Protein per pound'] = (df['Total protien'] / df['Price (£)']).round().astype(int)

df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
top_protein = df.sort_values('Protein per pound', ascending=False)
plt.barh(top_protein['Food'], top_protein['Protein per pound'])
plt.xlabel('Protein per £ (grams)')
plt.title('Best Protein Value Foods')
plt.show()
```
